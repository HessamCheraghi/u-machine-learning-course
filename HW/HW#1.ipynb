```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

data = pd.read_csv("HW#1StudentsPerformance.csv")

data
```

```python
data.describe()
```

```python
x = data['math score']
y = data['writing score']

plt.style.use('dark_background')
plt.scatter(x, y)
plt.xlabel('math score')
plt.ylabel('writing score')
plt.title("Writing score based on Math score")
plt.show()
```

```python

model = sm.OLS(y,sm.add_constant(x)).fit()
model.summary()
```

```python
plt.style.use('dark_background')
plt.scatter(x, y)


x_coef = model.params['math score']
const = model.params['const']
y_hat = x_coef * x + const
fig = plt.plot(x,y_hat,  lw=3, c='hotpink')

plt.xlabel('math score')
plt.ylabel('writing score')
plt.title("Writing score based on Math score")

plt.show()
```
